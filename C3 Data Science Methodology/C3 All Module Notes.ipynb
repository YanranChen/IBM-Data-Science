# Data Science Methodology 

## Overview
A Methodology is:
- A system of methods
- A guideline for decision making during the scientific process

Data Science Methodology guides data scientists in solving complex problems with data.

Using a methodology: 
- provides practical guidence
- avoids the mistakes that can happen by jumping to solutions before the analysis

## Data Science Methodology
by John Rollins
#### 1. Business understanding
*What is the problem that you are trying to solve?*

Every project, regardless of its size, starts with business understanding, which lays the foundation for successful resolution of the business problem. The business sponsors needing the analytic solution play the critical role in this stage by; Defining the problem, defining the project objectives and defining the solution requirements from a business perspective.

#### 2. Analytics approach
*How can the data be used to answer the question?*

After clearly stating a business problem, the analytic approach to solving the problem can be defined. Doing so involves expressing the problem in the context of statistical and analytical techniques so that techniques suitable for achieving the desired outcome can be identified. For example; If the question is to show relationships, then a descriptive approach would be applicable.

#### 3. Data requirements
*What data do you need to answer the question?*

Choice of analytic approach determines the data requirements, the analytic methods to be used require particular data content, formats and representations, guided by domain knowledge.

#### 4. Data collection
*Where is the data coming from (Identify all sources) and how will you get it?*

The data scientist identifies and gathers data resources—structured, unstructured and semi-structured—that are relevant to the problem domain. On encountering gaps in data collection, the data scientist might need to revise the data requirements and collect more data.

#### 5. Data understanding
*Is the data that you collected representative of the problem to be solved?*

Descriptive statistics and visualization techniques can help a data scientist understand data content, assess data quality and discover initial insights into the data. A revisiting of the previous step, data collection, might be necessary to close gaps in understanding.

#### 6. Data preparation
*What additional work is required to manipulate and work with the data?*

Data preparation is the most time-consuming phase of a data science project. Transforming data in the data preparation phase is the process of getting the data into a state where it is easier to work with. To work effectively with data, it must be prepared in a way that addresses missing or invalid values, removes duplicates and ensuring the data is properly formatted.

#### 7. Modeling
*In what way can the data be visualized to address the business problem?*

Data modeling focuses on developing models that are either descriptive or predictive. These models are based on the analytic approach that was taken, either statistical or machine learning driven.

#### 8. Evaluation
*Does the model used really answer the initial question or does it need to be adjusted?*

Model evaluation is performed during model development before deployment. Evaluation helps to assess the quality of the model and if the model meets the initial requirements.

#### 9. Deployment
*Can the model be put into practice?*

Once the model is developed and evaluated, the model is deployed and tested by the users.

#### 10. Feedback
*Can you get constructive feedback from the data and the stakeholder to answering the business question?*

Once in play, feedback from the users will help to refine the model and assess it for performance and impact.

## Analytics Approach
The process of selecting the appropriate method or path to address a specific data science question or problem.
### 1. Descriptive questions:
*What is the current status?*
- **Approach:** Descriptive Analytics
- **Techniques:**
  - Data aggregation
  - Data mining
  - Data visualization


### 2. Diagnostic questions:
*Why did it happen?*
- **Approach:** Diagnostic Analytics
- **Techniques:**
  - Drill-down: Exploring detailed data to find underlying causes.
  - Data discovery: Identifying patterns and relationships in data.
  - Correlation analysis: Assessing the relationship between different variables.

### 3. Predictive questions:
*What is likely to happen?*
- **Approach:** Predictive Analytics
- **Techniques:**
  - Regression analysis
  - Time series forecasing
  - Maching leaning models
 
### 4. Prescriptive questions:
*What should we do?*
- **Approach:** Prescriptive Analytics
- **Techniques:**
  - Optimization models: Finding the best solution from a set of alternatives.
  - Simulation: Modeling scenarios to predict outcomes.
  - Decision analysis: evaluating and comparing different decisions/

### 5. Classification questions:
*Which category does this data point belong to?*
- **Approach:** Classification
- **Techniques:**
  - Logistic regression
  - Decision trees
  - Support vector machines
  - Neural networks
